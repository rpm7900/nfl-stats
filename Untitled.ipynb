```python
import pandas as pd 
```

```python
# read csv into pandas 
stats= pd.read_csv("resources/Basic_Stats.csv")
stats_pd = pd.DataFrame(stats)
stats_pd
```

```python
# split and rearrange names in name column
stats_pd['Name'] = stats_pd['Name'].str.split(', ').str[::-1].str.join(' ')
stats_pd_mod = pd.DataFrame(stats_pd)
```

```python
# rename columns 
stats_renamed =stats_pd_mod.rename(columns ={'Age':'age', 'Birth Place':'birth_place','Birthday':'birthday',
                              'College':'college','Current Status':'current_stat','Current Team':'current_team',
                              'Current Team':'current_team','Experience':'experience','Height (inches)': 'height_in',
                              'High School': 'high_school','High School Location': 'high_school_loc','Name':'name',
                              'Number':'number','Player Id':'player_id','Position':'position','Weight (lbs)':'weight_lbs',
                               'Years Played':'yrs_played'})
```

```python
# rearranged columns 
stats_rearranged =stats_renamed[['name','age','birth_place','birthday','weight_lbs','height_in',
               'college','player_id','number','position','current_stat','current_team',
               'experience','high_school','high_school_loc','yrs_played']]
stats_cleaned = stats_rearranged.reset_index(drop = True)
```

```python
# read df back into csv
stats_cleaned.to_csv("resources/player_stats.csv",index = False,index_label = False)
```

```python
# Read in player salary CSV
player_salary = pd.read_csv("resources/NFL2017PlayerSalaries.csv")
player_salary
```
